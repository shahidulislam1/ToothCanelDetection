```python
from matplotlib import pyplot as plt
import cv2
import os
from random import choice
import shutil
%matplotlib inline
```

```python
#arrays to store file names
imgs =[]
xmls =[]

#setup dir names
# trainPath = 'D:/yolov5_traffic_sign_detection/dataset/images/train'
# valPath = 'D:/yolov5_traffic_sign_detection/dataset/images/val'
crsPath = '/home/tareq/Downloads/CSE465/valimage_augment_yolov5v3/' #dir where images and annotations stored

#setup ratio (val ratio = rest of the files in origin dir after splitting into train and test)
# train_ratio = 0.65
# val_ratio = 0.15
# test_ratio=0.20


#total count of imgs
totalImgCount = len(os.listdir(crsPath))/2

#soring files to corresponding arrays
for (dirname, dirs, files) in os.walk(crsPath):
    for filename in files:
        if filename.endswith('.txt'):
            xmls.append(filename)
        else:
            imgs.append(filename)
print(len(imgs))
```

```python
crsPath = '/home/tareq/Downloads/CSE465/valimage_augment_yolov5v3/'
trainpath0="/home/tareq/Downloads/CSE465/resnet50image/val/0"
trainpath1="/home/tareq/Downloads/CSE465/resnet50image/val/1"
trainpath2="/home/tareq/Downloads/CSE465/resnet50image/val/2"
c0=0
c1=0
c2=0
for f in range(len(imgs)):
    fileJpg = imgs[f] # get name of random image from origin dir
    fileXml = fileJpg[:-4] +'.txt' # get name of corresponding annotation file
    imagePath="/home/tareq/Downloads/CSE465/valimage_augment_yolov5v3/"+fileJpg
    labelPath="/home/tareq/Downloads/CSE465/valimage_augment_yolov5v3/"+fileXml
    image = cv2.imread(imagePath)
    (hI, wI) = image.shape[:2]
    lines = [line.rstrip('\n') for line in open(labelPath)]
            #if(len(objects)<1):
            #    raise Exception("The xml should contain at least one object")
    boxes = []
    if lines != ['']:
        for line in lines:
            components = line.split(" ")
            category = components[0]
            
            x  = int(float(components[1])*wI - float(components[3])*wI/2)
            y = int(float(components[2])*hI - float(components[4])*hI/2)
            h = int(float(components[4])*hI)
            w = int(float(components[3])*wI)
            imgCrop = image[y:y+h,x:x+w]
            resized_image = cv2.resize(imgCrop, (224, 224)) 
            if str(category) == "0":
                print("category0"+str(f))
                cv2.imwrite(trainpath0+"/"+str(f)+"_"+category+"_"+str(x)+"_"+fileJpg, resized_image)
                c0=c0+1
            elif category == "1":
                print("category1"+str(f))
                cv2.imwrite(trainpath1+"/"+str(f)+"_"+category+"_"+str(x)+"_"+fileJpg, resized_image)
                c1=c1+1
            elif category == "2":
                print("category2"+str(f))
                cv2.imwrite(trainpath2+"/"+str(f)+"_"+category+"_"+str(x)+"_"+fileJpg, resized_image)
                c2=c2+1
```
